```python
# 오프라인 문화활동 : 기준 - 문화활동, 영화, 자연, 축제 등 관람
# 개인영화관람수, 개인생활공예활동수, 개인자연명소관람수, 개인전시관람수,
# 개인노래방방문수, 개인사회봉사활동수, 개인전통예술학습활동수, 개인문화유적방문수,개인문학행사참여수, 
# 개인테마파크놀이공원동물원식물원방문수, 개인박물관관람수, 개인지역축제참가수, 개인전통예술공연관람수
data_off =  data[(data.Category == "개인영화관람수") | 
                (data.Category == "개인미술활동수") | (data.Category == "개인생활공예활동수") |
                (data.Category == "개인사진촬영수") |(data.Category == "개인자연명소관람수") | 
                (data.Category == "개인전시관람수") | (data.Category == "개인전통예술학습활동수") | 
                (data.Category == "개인문화유적방문수")|(data.Category == "개인테마파크놀이공원동물원식물원방문수") | 
                (data.Category == "개인박물관관람수") | (data.Category == "개인지역축제참가수") | 
                (data.Category == "개인전통예술공연관람수") | (data.Category == "개인문학행사참여수") | 
                (data.Category == "개인국내캠핑활동수")] 
data_off = data_off.sort_values(by = "Value", ascending = False)

plt.figure(figsize=(20,15))
sns.barplot(y="Category", x="Value", data=data_off)
plt.xlabel("활동수", fontsize = 16)
plt.ylabel("활동명", fontsize = 16)
plt.rc('ytick', labelsize=15) 
plt.rc('xtick', labelsize=15) 
plt.title("국민여가활동 - 문화생활", fontsize = 23) 
```
